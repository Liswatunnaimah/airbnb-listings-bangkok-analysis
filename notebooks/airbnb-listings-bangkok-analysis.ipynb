````python
### Cek Jumlah Missing Value per Kolom (Setelah Drop Kolom Tidak Relevan)

# Hitung jumlah missing value dan persentasenya
missing_values = df.isna().sum()
missing_percentage = (missing_values / len(df)) * 100

# Buat DataFrame untuk merangkum hasilnya
missing_df = pd.DataFrame({
    'Total Missing': missing_values,
    'Persentase (%)': missing_percentage
}).sort_values(by='Persentase (%)', ascending=False)

# Tampilkan hasil
missing_df
````

````python
### Cek Apakah Nilai Kosong di 'last_review' & 'reviews_per_month' Hanya Terjadi Saat number_of_reviews = 0

# Subset data yang memiliki missing value
missing_reviews_per_month = df[df['reviews_per_month'].isna()]
missing_last_review = df[df['last_review'].isna()]

# Cek apakah semua baris yang missing terjadi saat number_of_reviews = 0
is_reviews_pm_missing_only_when_no_reviews = (missing_reviews_per_month['number_of_reviews'] == 0).all()
is_last_review_missing_only_when_no_reviews = (missing_last_review['number_of_reviews'] == 0).all()

# Cetak hasil
print("Apakah 'reviews_per_month' hanya missing ketika 'number_of_reviews' = 0?")
print(is_reviews_pm_missing_only_when_no_reviews)

print("\nApakah 'last_review' hanya missing ketika 'number_of_reviews' = 0?")
print(is_last_review_missing_only_when_no_reviews)
````

## Penanganan Missing Value Berdasarkan Tujuan Analisis

### Tujuan Analisis
> Mengidentifikasi faktor-faktor yang berkontribusi terhadap **popularitas listing Airbnb di Bangkok**, dengan metrik utama seperti:
- `number_of_reviews`
- `reviews_per_month`
- `availability_365`

---

### Strategi Penanganan Missing Value

| Kolom                | % Missing | Fungsi Analisis         | Penanganan         | Alasan                                                                 |
|----------------------|-----------|--------------------------|--------------------|------------------------------------------------------------------------|
| `reviews_per_month`  | 36.5%     | Indikator popularitas   | **Imputasi 0**     | Nilai 0 merepresentasikan listing yang tidak aktif / belum ada review. |
| `last_review`        | 36.5%     | Indikator aktivitas     | **Biarkan NaT**    | Berguna untuk klasifikasi listing yang tidak aktif atau baru dibuat.  |

# **OUTLIER**

````python
# Cek nilai outlier pakai IQR
# Fungsi untuk mendeteksi outlier menggunakan metode IQR
def detect_outlier_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    
    print(f"\n--- Outlier Detection for '{col}' ---")
    print(f"Q1: {Q1}")
    print(f"Q3: {Q3}")
    print(f"IQR: {IQR}")
    print(f"Batas bawah (lower): {lower}")
    print(f"Batas atas (upper): {upper}")
    print(f"Jumlah outlier: {len(outliers)}")
    print("-" * 50)

# Kolom numerik yang ingin dicek
cols_to_check = ['price', 'minimum_nights', 'number_of_reviews']
for col in cols_to_check:
    detect_outlier_iqr(df, col)
````

## Analisis : Regresi Poisson untuk Menilai Faktor-Faktor Penentu Popularitas

### Tujuan Analisis
- Mengevaluasi pengaruh simultan dari berbagai faktor terhadap **popularitas listing** yang diukur melalui **jumlah review (number_of_reviews)**. Analisis ini dapat digunakan untuk:
- Menilai kekuatan pengaruh tiap variabel seperti tipe kamar, harga, ketersediaan, dan minimum stay.
- Menyusun strategi berbasis data untuk host dan platform.

### Alasan Pemilihan Model: GLM Poisson
- Variabel target `number_of_reviews` adalah data **count** (bilangan bulat, tidak negatif).
- Distribusi skewed ke kanan → tidak cocok pakai OLS (regresi linear biasa).
- **GLM dengan distribusi Poisson dan link function log** adalah pendekatan yang **ideal untuk memodelkan data count**.

### Spesifikasi Model
**Model:** Generalized Linear Model (Poisson Regression)

**Formula:**
```
log(E[number_of_reviews]) = β0 + β1*C(room_type) + β2*price + β3*availability_365 + β4*minimum_nights
```

### Variabel
| Jenis | Nama Variabel | Tipe | Keterangan |
|-------|----------------|------|------------|
| Target | number_of_reviews | Count | Total ulasan sebagai indikator popularitas |
| Prediktor | room_type | Kategorikal | Tipe kamar (Entire, Private, Shared, Hotel) |
| Prediktor | price | Numerik | Harga listing per malam (dalam Baht) |
| Prediktor | availability_365 | Numerik | Jumlah hari listing tersedia dalam 1 tahun |
| Prediktor | minimum_nights | Numerik | Minimum jumlah malam untuk menginap |

### Statistik Uji dan Hipotesis
#### Jenis Uji:
- Wald Test (Uji Z) digunakan untuk mengetahui apakah masing-masing koefisien dalam model signifikan atau tidak.
- Penilaian dilakukan melalui Z-score dan p-value dari hasil model.summary().

#### Hipotesis Statistik:
| H0 | Koefisien = 0 (tidak ada pengaruh signifikan) |
|----|--------------------------------------------|
| H1 | Koefisien ≠ 0 (ada pengaruh signifikan)    |

#### Kriteria Uji:
- Jika **p-value < 0.05** → variabel signifikan
- Jika **p-value ≥ 0.05** → variabel tidak signifikan

---

### Fit Model GLM Poisson

````python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat model regresi Poisson
model_poisson = smf.glm(
    formula='number_of_reviews ~ C(room_type) + price + availability_365 + minimum_nights',
    data=df,
    family=sm.families.Poisson()
).fit()

# Tampilkan ringkasan model
print(model_poisson.summary())
````

### Interpretasi GLM Poisson 
#### Rumusan Masalah: 
Apakah variabel room_type, price, availability_365, dan minimum_nights berpengaruh signifikan terhadap popularitas listing (diukur dari number_of_reviews)?

Hipitesis:
- H0 (hipotesis nol): Koefisien variabel = 0 (tidak berpengaruh signifikan terhadap jumlah review)
- H1 (hipotesis alternatif): Koefisien variabel ≠ 0 (berpengaruh signifikan)
## Hasil Uji Statistik Inferensia

### Model: Generalized Linear Model (GLM) – Poisson Regression

**Tujuan:** Mengidentifikasi faktor-faktor yang berkontribusi terhadap jumlah review (sebagai indikator popularitas listing Airbnb).

**Hipotesis Umum:**
- H0 (null): Variabel-variabel independen tidak berpengaruh signifikan terhadap jumlah review.
- H1 (alternatif): Terdapat pengaruh signifikan dari variabel-variabel terhadap jumlah review.

### Variabel yang Diujikan:
1. **Tipe kamar (room_type)**
2. **Harga per malam (price)**
3. **Jumlah hari tersedia dalam setahun (availability_365)**
4. **Durasi minimum menginap (minimum_nights)**

### Hasil Uji dan Interpretasi

  - Menunjukkan baseline jumlah review untuk `Entire home/apt`.

- **room_type: Hotel room**
  - Signifikan (p < 0.001)
  - Jumlah review lebih rendah sekitar 13–14% dibanding `Entire home/apt`.

- **room_type: Private room**
  - Signifikan (p < 0.001)
  - Review turun ~60% dibanding entire home/apt. Tamu cenderung menghindari berbagi ruang.

  - Signifikan (p < 0.001)
  - Menarik: review menurun ketika listing terlalu "tersedia" → mungkin tidak menarik atau sepi peminat.

- **minimum_nights**

### Insight & Rekomendasi:

# **Korelasi Ketersediaan Listing dan Jumlah Review**

### **Apakah availability_365 (ketersediaan listing dalam setahun) berpengaruh terhadap popularitas listing (jumlah review)?**

**Hipotesis:**
- H0 (null): Tidak ada perbedaan signifikan jumlah review berdasarkan level availability.
- H1 (alternatif): Ada perbedaan signifikan jumlah review berdasarkan level availability.

**Langkah Analisis:**
- Visualisasi Korelasi
- Plot Boxplot: Jumlah Review vs Availability Group
- Uji Normalitas : untuk liat datanay normal atau ga kalo berdist normal pake ANOVA, kalo ga pake Kruskal-Wallis
- Uji Statistik:Kruskal-Wallis krena data ga bersidtribusi normal dan terdiri dari > 2 grup.

---

#### Langkah 1: Buat Grup Ketersediaan
- Tujuan : mengelompokkan listing berdasarkan jumlah hari tersedia dalam setahun (availability_365)

````python
# Bikin grup availability agar mudah dibandingkan
df['availability_group'] = pd.cut(df['availability_365'],
                                   bins=[-1, 0, 100, 200, 300, 365],
                                   labels=['0', '1-100', '101-200', '201-300', '301-365'])
````

#### Langkah 2: Visualisasi Boxplot

````python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='availability_group', y='number_of_reviews')
plt.title("Jumlah Review per Grup Availability")
plt.xlabel("Availability (Hari)")
plt.ylabel("Jumlah Review")
plt.tight_layout()
plt.show()
````

### Interpretasi Boxplot number_of_reviews vs availability_group
- Semakin tinggi ketersediaan (availability), cenderung memiliki median review lebih tinggi.
- Grup 301–365 hari menunjukkan sebaran jumlah review yang paling besar → indikasi bahwa listing yang selalu aktif punya potensi popularitas lebih tinggi.
- Grup 0 hari tetap bisa punya review, tapi median sangat rendah → mungkin karena listing sebelumnya aktif tapi kini sudah tidak tersedia.
- Terdapat outlier di semua grup, menunjukkan bahwa popularitas tidak hanya bergantung pada availability saja (ada faktor lain juga seperti lokasi, harga, dll).

 Insight: Ketersediaan yang tinggi bisa jadi salah satu indikator popularitas, tapi bukan satu-satunya faktor.

 ---

#### Langkah 3: Uji Normalitas

````python
# Cek distribusi normal dulu

from scipy.stats import shapiro

for group in df['availability_group'].unique():
    stat, p = shapiro(df[df['availability_group'] == group]['number_of_reviews'])
    print(f"{group} - p-value: {p:.4f}")
````

#### Visualisasi distribusi

````python
for group in df['availability_group'].unique():
    plt.figure(figsize=(6, 3))
    subset = df[df['availability_group'] == group]['number_of_reviews']
    sns.histplot(subset, kde=True, bins=30)
    plt.title(f'Distribusi number_of_reviews - Group {group}')
    plt.xlabel('Number of Reviews')
    plt.tight_layout()
    plt.show()
````

### Interpretasi Hasil Uji Normalitas & Visualisasi
1. Hasil Shapiro-Wilk Test (uji normalitas):
    - Semua grup availability_group memiliki p-value = 0.0000, artinya:
    - Distribusi jumlah review tidak normal untuk semua grup.

2. Visualisasi histogram per grup juga menunjukkan:
    - Distribusi miring ke kanan (right-skewed).
    - Banyak data mendekati nol, dengan outlier di sisi kanan.
    - Mendukung hasil uji Shapiro-Wilk → distribusi tidak normal secara visual.

---

#### Langkah 4 : Uji Kruskal-Wallis Test (alternatif non-parametrik dari ANOVA)
- karena data ga berdistribusi normal dan terdiri dari > 2 grup, maka pake Kruskal-Wallis Test
- Fungsi: menguji apakah median jumlah review antar grup availability berbeda signifikan
- Cocok karena:
    - Data numerik (jumlah review)
    - Grup kategorikal (> 2: ada 5 grup availability)
    - Tidak memenuhi asumsi normalitas dan homogenitas varians

---

````python
from scipy.stats import kruskal

# Ambil data berdasarkan grup availability
groups = [df[df['availability_group'] == group]['number_of_reviews'] for group in df['availability_group'].unique()]

# Uji Kruskal-Wallis
stat, p = kruskal(*groups)
print(f"Hasil Uji Kruskal-Wallis: Statistik = {stat:.4f}, p-value = {p:.4f}")
````

### Interpretasi Apakah availability_365 (ketersediaan listing dalam setahun) berpengaruh terhadap popularitas listing (jumlah review):
- p-value < 0.05 → artinya kita menolak H0.

# **Room Type vs Jumlah Review**
### **Analisis Pengaruh Room Type terhadap Popularitas Listing Airbnb di Bangkok**

### Tujuan Analisis
Menilai apakah **tipe kamar (room_type)** berpengaruh terhadap **popularitas listing**, yang diukur dari **jumlah review (number_of_reviews)**.

### Rumusan Masalah
Apakah terdapat perbedaan signifikan jumlah review berdasarkan tipe kamar di Airbnb Bangkok?

### Hipotesis 
- **H0 (Hipotesis Nol):** Tidak ada perbedaan jumlah review rata-rata antar tipe kamar.
- **H1 (Hipotesis Alternatif):** Ada perbedaan jumlah review rata-rata antar tipe kamar.

---

Langkah 1: Visualisasi Boxplot (EDA)

````python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(x='room_type', y='number_of_reviews', data=df)
plt.title('Distribusi Jumlah Review Berdasarkan Tipe Kamar')
plt.xlabel('Room Type')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
````

### Interpretasi Visualisasi Boxplot
- Visualisasi memperlihatkan perbedaan yang cukup mencolok antar tipe kamar.
- Entire home/apt memiliki persebaran jumlah review yang paling tinggi.
- Shared room dan Hotel room menunjukkan persebaran review yang lebih rendah.
- Insight Awal: Ada indikasi perbedaan popularitas berdasarkan tipe kamar.



Langkah 2: Uji Normalitas per Kelompok (Shapiro-Wilk)
- Jika p-value < 0.05 untuk semua grup → data tidak normal → lanjut ke Kruskal-Wallis.

````python
from scipy.stats import shapiro

for room in df['room_type'].unique():
    subset = df[df['room_type'] == room]['number_of_reviews']
    stat, p = shapiro(subset)
    print(f"{room} - p-value: {p:.4f}")
````

### Intepretasi Uji Normalitas (Shapiro-Wilk)
- Semua kelompok room_type memiliki p-value < 0.05.
- Artinya, data tidak berdistribusi normal.
- Karena data tidak normal dan jumlah grup > 2, maka tidak cocok menggunakan ANOVA.
- Lanjut ke uji non-parametrik: Kruskal-Wallis.

Langkah 3: Uji Statistik Inferensia (Kruskal-Wallis)
- Jika p-value < 0.05 → lanjut ke uji post-hoc.

````python
from scipy.stats import kruskal

groups = [df[df['room_type'] == room]['number_of_reviews'] for room in df['room_type'].unique()]
stat, p = kruskal(*groups)

print(f"Hasil Uji Kruskal-Wallis: Statistik = {stat:.4f}, p-value = {p:.4f}")
````

### Intepretasi Uji Kruskal-Wallis
- Hasil: p-value = 0.0000
-  Interpretasi:
    - Menolak H0.
    - Ada perbedaan signifikan jumlah review berdasarkan room_type.
- Namun, uji ini tidak memberitahu pasangan mana yang beda signifikan -> lanjut ke post-hoc.



Langkah 4: Post-Hoc Test (Dunn’s Test)

````python
import scikit_posthocs as sp

# Pastikan df tidak memiliki NaN di kolom yang digunakan
posthoc_df = df[['room_type', 'number_of_reviews']].dropna()

# Uji Dunn dengan koreksi Bonferroni
sp.posthoc_dunn(posthoc_df, val_col='number_of_reviews', group_col='room_type', p_adjust='bonferroni')
````

### Interpretasi Post-hoc Dunn’s Test (Bonferroni Correction)
- Semua perbandingan antar tipe kamar signifikan (p < 0.05), kecuali:
    - Hotel room vs Shared room → tidak signifikan (p > 0.05)
- Entire home/apt unggul signifikan dibanding semua tipe lain.

---

### Kesimpulan & Rekomendasi
- Room_type secara signifikan memengaruhi jumlah review.
- Entire home/apt adalah tipe dengan jumlah review tertinggi → menunjukkan preferensi tamu terhadap privasi penuh.
- Untuk host baru, disarankan memilih listing dengan tipe Entire home/apt untuk memaksimalkan visibilitas dan potensi review.

# **Segmentasi Listing Populer vs Tidak Populer**
 - Segmentasi berdasarkan Q1 dan Q3 dari `number_of_reviews`

### Tujuan Analisis:
- Mengelompokkan listing menjadi:
- Low Popularity (jumlah review di bawah atau sama dengan Q1)
- High Popularity (jumlah review di atas atau sama dengan Q3)
- Membandingkan karakteristik kedua kelompok tersebut agar host bisa memahami perbedaan listing populer dan tidak populer.


### Segmentasi Listing Populer vs Tidak Populer

````python
import pandas as pd

# Hitung Q1 dan Q3 dari jumlah review
q1 = df['number_of_reviews'].quantile(0.25)
q3 = df['number_of_reviews'].quantile(0.75)

# Buat kolom kategori popularitas
def segment_popularity(x):
    if x <= q1:
        return 'Low Popularity'
    elif x >= q3:
        return 'High Popularity'
    else:
        return 'Mid'

df['popularity_segment'] = df['number_of_reviews'].apply(segment_popularity)

# Cek distribusi segmen
print(df['popularity_segment'].value_counts())
````

### Interpretasi Singkat: Segmentasi Listing Berdasarkan Popularitas
- Segmentasi dilakukan berdasarkan kuartil jumlah review (Q1 dan Q3):
    - Low Popularity: listing dengan jumlah review ≤ Q1 (kuartil bawah)
    - High Popularity: listing dengan jumlah review ≥ Q3 (kuartil atas)
    - Mid: sisanya di antara Q1 dan Q3
- Segmentasi ini bertujuan untuk memetakan perbedaan karakteristik antara listing populer dan tidak populer.

### Insight 
- Data terdistribusi cukup seimbang antar segmen karena berbasis kuartil (25%-50%-25%).

## **Analisis Tambahan : Pengaruh Penggunaan Kata Kunci dalam Nama Listing terhadap Popularitas Listing Airbnb Bangkok**

### Tujuan Analisis
Menilai apakah penggunaan kata kunci tertentu dalam nama listing (seperti "BTS", "Pool", "Wifi", dll) berpengaruh terhadap **popularitas listing**, yang diukur dari **jumlah ulasan (number_of_reviews)**.

### Rumusan Masalah
Apakah terdapat perbedaan jumlah review antara listing yang mengandung kata kunci tertentu di nama listing dibandingkan dengan yang tidak?

### Hipotesis Statistik
- **H0 (Hipotesis Nol):** Rata-rata jumlah review **tidak berbeda** secara signifikan antara listing dengan nama yang mengandung kata kunci dan yang tidak.
- **H1 (Hipotesis Alternatif):** Rata-rata jumlah review **berbeda** secara signifikan antara listing dengan nama yang mengandung kata kunci dan yang tidak.

### Uji Statistik yang Digunakan
- **Jenis uji:** Uji t dua sampel independen (Independent Two-Sample t-Test)
- **Alasan:** Variabel independen adalah kategorikal biner (mengandung kata kunci atau tidak), dan variabel dependen adalah numerik (jumlah review).
- **Kriteria pengambilan keputusan:**
  - Jika p-value < 0.05 → Tolak H0 → Ada perbedaan signifikan
  - Jika p-value ≥ 0.05 → Gagal tolak H0 → Tidak ada perbedaan signifikan

---

### Langkah 1: Identifikasi Kata-Kata Populer dalam Nama Listing
- Tujuan: Menemukan kata kunci yang sering muncul agar bisa dijadikan acuan.
- Penjelasan:
Kita akan gabungkan semua nama, ubah jadi huruf kecil, lalu hitung frekuensi setiap kata.

````python
from collections import Counter
import re

# Gabungkan semua nama listing dan ubah jadi lowercase
all_words = ' '.join(df['name'].dropna().str.lower())

# Split dan hitung frekuensi kata
word_list = re.findall(r'\w+', all_words)
word_counts = Counter(word_list)

# Tampilkan 30 kata paling sering muncul
print(word_counts.most_common(30))
````

### Interpretasi Identifikasi Kata Populer
- Hasil dari word_counts.most_common() menunjukkan kata-kata yang paling sering muncul di nama listing seperti:
'bts', 'wifi', 'pool', 'view', 'balcony', 'skytrain', dll.
- Kata-kata ini sangat mungkin menarik perhatian calon tamu karena berhubungan dengan lokasi strategis (BTS, Skytrain), fasilitas (pool, wifi), dan view.
- Insight: Kata-kata ini layak dijadikan indikator karena mewakili nilai jual dari sebuah listing.

### Langkah 2: Buat Kolom has_keyword
- Tujuan: Tandai listing yang mengandung salah satu kata kunci populer.
- Penjelasan: Kita akan cek apakah nama listing mengandung salah satu kata dari daftar keyword.

````python
# Kata kunci yang dipilih berdasarkan konteks dan popularitas
keywords = ['bts', 'wifi', 'pool', 'city', 'view', 'balcony', 'skytrain']

# Fungsi untuk mendeteksi apakah nama mengandung kata kunci
def contains_keyword(name):
    if pd.isna(name):
        return 0
    name_lower = name.lower()
    return int(any(kw in name_lower for kw in keywords))

# Tambahkan kolom biner
df['has_keyword'] = df['name'].apply(contains_keyword)

# Cek distribusinya
print(df['has_keyword'].value_counts())
````

### Intepretasi Kolom has_keyword
- Ditemukan bahwa:
    - 6.129 listing menggunakan kata kunci.
    - 9.716 listing tidak menggunakan kata kunci.

- Insight:
Masih banyak host belum memanfaatkan strategi pemilihan nama yang menarik, padahal itu bisa menjadi nilai jual.

### Langkah 3: Visualisasi Jumlah Review Berdasarkan has_keyword
- Tujuan: Melihat secara visual apakah ada perbedaan jumlah review antara listing yang pakai keyword dan yang tidak.

````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.boxplot(x='has_keyword', y='number_of_reviews', data=df)
plt.title('Jumlah Review Berdasarkan Penggunaan Kata Kunci di Nama Listing')
plt.xlabel('Mengandung Kata Kunci (1 = Ya, 0 = Tidak)')
plt.ylabel('Jumlah Review')
plt.show()
````

### Intepretasi Visualisasi Boxplot
- Median jumlah review lebih tinggi pada listing yang menggunakan kata kunci.
- Terdapat lebih banyak outlier positif (jumlah review tinggi) di grup has_keyword = 1.

- Insight:
Secara visual, ada indikasi bahwa listing dengan keyword memang cenderung lebih populer.

### Langkah 4: Uji Normalitas
- Tujuan: Menentukan uji statistik yang tepat berdasarkan distribusi data.
-  Jika p-value < 0.05 → distribusi tidak normal → pakai Mann-Whitney U Test.

````python
from scipy.stats import shapiro

group0 = df[df['has_keyword'] == 0]['number_of_reviews']
group1 = df[df['has_keyword'] == 1]['number_of_reviews']

# Uji Shapiro-Wilk untuk normalitas
print("Grup tanpa keyword:", shapiro(group0))
print("Grup dengan keyword:", shapiro(group1))
````

### Interpretasi Uji Normalitas (Shapiro-Wilk)
- Hasil:
    - Grup 0 (tanpa keyword): p-value ≈ 0.0
    - Grup 1 (dengan keyword): p-value ≈ 0.0
- Interpretasi: Distribusi data pada kedua grup tidak normal, maka tidak cocok pakai uji t, dan dilanjut ke Mann-Whitney U Test.

Langkah 5: Uji Statistik (Mann-Whitney U Test)
Tujuan: Menguji apakah perbedaan jumlah review antara dua grup signifikan.

````python
from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(group0, group1, alternative='two-sided')
print(f"Hasil Uji Mann-Whitney: Statistik = {stat:.4f}, p-value = {p:.4f}")
````

### Interpretasi Uji Statistik (Mann-Whitney U Test) 
Karena p-value < 0.05, maka tolak H0
- Ada perbedaan signifikan jumlah review antara listing yang menggunakan kata kunci dan yang tidak.

## Insight Keseluruhan
Berdasarkan seluruh rangkaian analisis, dapat disimpulkan bahwa listing Airbnb yang paling populer di Bangkok menunjukkan karakteristik sebagai berikut:
- Menggunakan **Entire home/apartment** sebagai room type
- Memiliki **harga kompetitif**, umumnya antara **900–2.500 Baht**
- Menetapkan **minimum stay fleksibel** (1–2 malam)
- Memiliki **availability tinggi** (aktif sepanjang tahun)
- Berlokasi di **distrik wisata utama** seperti Khlong Toei, Watthana, dan Sathorn
- Menggunakan **kata kunci strategis** pada nama listing ("Wifi", "BTS", "Central")

Strategi kombinasi dari elemen-elemen ini meningkatkan peluang listing untuk lebih sering muncul di hasil pencarian, dipesan, dan mendapatkan review positif.

## Actionable Recommendations

### Untuk Host Baru
- Buat listing bertipe **Entire home/apt** agar menarik pasar luas.
- Fokus listing di area dengan traffic tinggi seperti **Khlong Toei**, **Watthana**, atau **Sathorn**.
- Tetapkan harga kompetitif (antara **900–2.500 Baht**), terutama untuk listing baru.
- Minimalkan **minimum stay** menjadi 1–2 malam agar lebih fleksibel.
- Tambahkan kata kunci penting di nama listing seperti "Near BTS", "Pool", "Central", dan "Wifi" untuk memaksimalkan visibilitas.

### Untuk Host Eksisting
- Evaluasi ulang harga listing jika sepi review dan sesuaikan dengan kategori harga wilayah.
- **Perbarui kalender secara berkala** untuk menjaga listing tetap aktif.
- Pertimbangkan **optimalisasi keyword** untuk meningkatkan Click-Through Rate (CTR).
- Bandingkan listing dengan power host dan pelajari apa yang membedakan mereka (foto, deskripsi, fasilitas, dan harga).

### Untuk Tim Manajemen Airbnb
- Optimalkan algoritma pencarian untuk memprioritaskan listing yang sesuai dengan pola sukses di atas.
- Kembangkan sistem **rekomendasi otomatis** kepada host berdasarkan performa historis data.
- Sediakan panduan strategi harga dan keyword berdasarkan data tiap distrik agar membantu host baru.
- Dorong host agar melakukan perbandingan harga dan keyword dengan power host di sekitarnya.
